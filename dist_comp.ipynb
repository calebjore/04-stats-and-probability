```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

print("Comparing the means of samples from the same distribution.\n")

x = np.random.normal(loc=0, scale=1, size=10)
y = np.random.normal(loc=0, scale=1, size=10)

t, p = stats.ttest_ind(x, y)

print(f"Sample 1 mean: {round(x.mean(), 3)}")
print(f"Sample 2 mean: {round(y.mean(), 3)}")

print(f"t-value: {round(t, 3)}")
print(f"p-value: {round(p, 3)}")

mu = 0
sigma = 1
dist_x = np.linspace(mu-3*sigma, mu+3*sigma, 1000)
dist_y = stats.norm.pdf(dist_x, mu, sigma)

fig, ax = plt.subplots()

ax.plot(dist_x, dist_y, label=r"$\mu=0$, $\sigma=1$")
ax.legend()
fig.suptitle("Distribution from which both samples were taken")
plt.show()
```

```python
print("Comparing the means of samples from different distributions.\n")

mu1 = 0
sigma1 = 1
mu2 = 2
sigma2 = 1

x = np.random.normal(loc=mu1, scale=sigma1, size=10)
y = np.random.normal(loc=mu2, scale=sigma2, size=10)

t, p = stats.ttest_ind(x, y)

print(f"Sample 1 mean: {round(x.mean(), 3)}")
print(f"Sample 2 mean: {round(y.mean(), 3)}")

print(f"t-value: {round(t, 3)}")
print(f"p-value: {round(p, 3)}")

x1 = np.linspace(mu1-3*sigma1, mu1+3*sigma1, 1000)
x2 = np.linspace(mu2-3*sigma2, mu2+3*sigma2, 1000)

probs1 = stats.norm.pdf(x1, mu1, sigma1)
probs2 = stats.norm.pdf(x2, mu2, sigma2)

fig, ax = plt.subplots()

ax.plot(x1, probs1, label=rf"$\mu={mu1}$, $\sigma={sigma1}$")
ax.plot(x2, probs2, label=rf"$\mu={mu2}$, $\sigma={sigma2}$")
ax.legend()

fig.suptitle("Distributions from which samples were taken")
plt.show()
```

```python
mu_norm = 0
sigma_norm = 1
num = 50

x_norm = np.random.normal(loc=mu_norm, scale=sigma_norm, size=num)
y_norm = np.random.normal(loc=mu_norm, scale=sigma_norm, size=num)
y_cauchy = np.random.standard_cauchy(size=num)

fig, axes = plt.subplots(1, 2, figsize=(10, 5), sharex=True, sharey=True)

axes[0].set_xlim(-10, 10)
axes[0].set_ylim(-10, 10)

axes[0].scatter(x_norm, y_norm)
axes[1].scatter(x_norm, y_cauchy)

axes[0].set_xlabel("normal")
axes[0].set_ylabel("normal")

axes[1].set_xlabel("normal")
axes[1].set_ylabel("standard Cauchy")

fig.suptitle("Scatterplot comparison of samples from \nnormal + normal and normal + Cauchy distributions")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5), sharex=True, sharey=True)

axes[0].set_xlim(-8, 8)
axes[0].set_ylim(-8, 8)

axes[0].scatter(np.sort(x_norm), np.sort(y_norm))
axes[1].scatter(np.sort(x_norm), np.sort(y_cauchy))

axes[0].set_xlabel("normal")
axes[0].set_ylabel("normal")

axes[1].set_xlabel("normal")
axes[1].set_ylabel("standard Cauchy")

fig.suptitle("Q-Q plots of data from \nnormal + normal and normal + Cauchy distributions")
plt.show()
```
