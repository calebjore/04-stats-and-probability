```python
import numpy as np
import matplotlib.pyplot as plt

# Simluating normal distribution with Monte Carlo method
fig, axes = plt.subplots(2, 2, sharex=True, figsize=(10,8))

n_vals = [100, 1000, 10000, 100000]

for ax, n in zip(axes.flatten(), n_vals):
    x = np.random.normal(loc=0, scale=1, size=n)
    ax.hist(x, bins=np.linspace(-3,3,121), label=f"n={n}")
    ax.legend()

fig.suptitle("Monte Carlo simulation of\nNormal Distribution")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, figsize=(10,8))

n_vals = [100, 1000, 10000, 100000]

for ax, n in zip(axes.flatten(), n_vals):
    x = np.random.exponential(scale=1, size=n)
    ax.hist(x, bins=np.linspace(-0,4,81), label=f"n={n}")
    ax.legend()

fig.suptitle("Monte Carlo simulation of\nExponential Distribution")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, figsize=(10,8))

n_vals = [100, 1000, 10000, 100000]

for ax, n in zip(axes.flatten(), n_vals):
    x = np.random.uniform(low=0, high=1, size=n)
    ax.hist(x, bins=np.linspace(-0,1,101), label=f"n={n}")
    ax.legend()

fig.suptitle("Monte Carlo simulation of\nUniform Distribution")
plt.show()
```
